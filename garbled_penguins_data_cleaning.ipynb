<h4>Question 3: Missing values</h4>
It seems like some of the researchers used SPSS and a standard where unknown values are coded as `-999.0`. 

Replace all the `-999.0` with the numpy variant of non-values, `np.nan`.

<h4>Question 4: Type conversion and inch to mm</h4>

It seems like the americans on Biscoe island recorded their measurements on bill depth in inches, while everyone else took proper measurements in mm. Somehow, it seems like this mistake also converted all the measurements to text strings with commas instead of dots. 

First of all, change all commas to dots in `bill_depth_mm`, then change the type to a numeric.

Then write a function `clean_bill_depth` that takes a single row of the dataframe and, if the value is from the island of `Biscoe`, it converts `bill_depth_mm` from inches to mm and returns it. Otherwise it returns `bill_depth_mm` as is. Then `apply` this function to convert all data into millimeters. 

1 inch is by definition exactly 25.4 mm.

```python
df['bill_depth_mm'] = df['bill_depth_mm'].replace(to_replace=',', value='.', regex=True)

df['bill_depth_mm'] = pd.to_numeric(df['bill_depth_mm'], errors='coerce')

def clean_bill_depth(row):
    # implement this function
    if row['island'] == 'Biscoe':
        return row['bill_depth_mm'] * 25.4
    else:
        return row['bill_depth_mm']

df['bill_depth_mm'] = df.apply(clean_bill_depth, axis=1)

df
```

<h4>Question 7: Custom summary function</h4>
Now the data is *finally* ready for some analysis.

Write a function that named `five_summary` that takes a list of numbers (we consider a Python list, a NumPy ndarray, or a Pandas series a list in this sense), and produces a dictionary as an output, containing the following fields:
- `minimum` containing the smallest value in the list,
- `maximum` containing the largest value in the list,
- `median` containing the median value of the list,
- `mean` containing the mean value of the list, and
- `sd` containing the sample standard deviation of the list. 

`Apply` this function to all all numeric columns of the dataset. Save the result in the variable `summary`. The result should be a `Series` that has the column names as the index and each element is a dictionary of the given form.

### Hints:
- The method `select_dtypes` of the dataframe can be used to select all columns of a specific datatype. `np.number` is a datatype that represents all numeric datatypes.
- Numpy comes with appropriate functions to compute these values, look them up!
- As there are missing values (NaNs), you need to handle these correctly (that is, ignore the values); NumPy provides variants of the basic functions for computing, e.g., mean, such that it ignores NaNs, look it up!
- For computing the standard deviation, use the argument `ddof=1` to specify 1 degree of freedom (this is because we use the sample mean to estimate the average; we will discuss why this correction is done later in class).
- The function `five_summary` should be able to handle also other data, as long as it is numerical; the automatic grader will test this on other data as well.

```python
def five_summary(column):
    column = np.array(column)
    
    summary = {
        'minimum': np.nanmin(column),  
        'maximum': np.nanmax(column),  
        'median': np.nanmedian(column),  
        'mean': np.nanmean(column),  
        'sd': np.nanstd(column, ddof=1)  
    }
    return summary

numeric_df = df.select_dtypes(include=[np.number])
    
summary = numeric_df.apply(five_summary, axis=0)
    
summary
```

<h4>Question 8: Grouped aggregate by the island</h4>

The result above is somewhat difficult to read and follow. Furthermore, since we are aggregating values from different populations (different species of penguins living on different island), it might not be the best course of action.

We are going to repeat this idea, but construct a hierarchical index where we group the penguins by the island they live in, then by the species, and then both. We are going to use the `agg` method to construct the same summary as before (`min`, `max`, `median`,`mean`,`std`). Group on island, species, and both. 

Save the answer in the variable `result_island`. The resulting dataframe should have a two-level hierarchical index where the first level corresponds to the column `island` and the second is labelled `variable` and contains the numerical column names. There should be 5 columns called `min`, `max`, `median`, `mean`, and `std`, and each row should contain these summary statistics for that particular column for those rows whose `island` corresponds to the value to the left.

That is, for example, if `island` is `Biscoe` and variable is `bill_depth_mm`, the value in column `mean` should be the mean of all `bill_depth_mm` values of those rows where `island` is `Biscoe`.

#### Hints:
- `pd.melt` is very useful, as we can then simply identify measurements by the species, island, sex, and variable in question, and this works nicely with `agg`.
- `drop` can be used to remove columns that are in the way for some operations.
- `groupby` is useful for creating the desired hierarchical index.
- `agg` can take a list of strings of the most commonly used operations.


```python
melted_df = pd.melt(df, id_vars=['island'], 
                    value_vars=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'year'],
                    var_name='variable', value_name='value')

grouped = melted_df.groupby(['island', 'variable']).agg(
    min=('value', 'min'),
    max=('value', 'max'),
    median=('value', 'median'),
    mean=('value', 'mean'),
    std=('value', 'std')
)

result_island = pd.concat({'value': grouped}, axis=1)
result_island 
```

<h4>Question 9: Grouped aggregate by the species</h4>
Do the same but group by the species. Save the answer in the variable `result_species`.

```python
melted_df = pd.melt(df, id_vars=['species'], 
                    value_vars=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'year'],
                    var_name='variable', value_name='value')

grouped = melted_df.groupby(['species', 'variable']).agg(
    min=('value', 'min'),
    max=('value', 'max'),
    median=('value', 'median'),
    mean=('value', 'mean'),
    std=('value', 'std')
)

result_species = pd.concat({'value': grouped}, axis=1)
result_species
```

<h4>Question 10: Grouped aggregate by the species and the island</h4>
Do the same but group by the species and island. The first level of the index should be species, the second island, and the third variable. Save the answer in the variable `result_both`.


```python
melted_df = pd.melt(df, id_vars=['species', 'island'], 
                    value_vars=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'year'],
                    var_name='variable', value_name='value')

grouped = melted_df.groupby(['species','island','variable']).agg(
    min=('value', 'min'),
    max=('value', 'max'),
    median=('value', 'median'),
    mean=('value', 'mean'),
    std=('value', 'std')
)

result_both = pd.concat({'value': grouped}, axis=1)
result_both
```
